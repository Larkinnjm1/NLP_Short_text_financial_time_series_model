Time series analysis section is broken down as the following:

The Exploratory Data Analysis (EDA) strategy taken was to divide the analysis of the news and stock indices data sets separately with the index data being analysed first with respect to time. This was done due to the fact that practically speaking analysing large quantities of text data has high time complexity which increases linearly with the number of topics and documents utilised in the model [6]. As a result, determining time periods of interest within the index dataset to analyse and then sampling from the news dataset for that period was performed in order to mitigate this constraint.

## Time Series Object Function set up 

Two objects where set up for time series data analysis:

- control_chart: The purpose of this object is to delineate between buy,sell and hold signals within the volatility trends for each index. Inputs incude the time series volatility data and the percent threshold for delineation between buy,sell and hold signals. 
- stat_significant: The purpose of this is to return a boolean to confirm whether or not a statistical result is significant or not. 

```python
#Function to generate buy sell or hold classification labels for each 
def control_chart(df,df_series,percent=0.01):

    blk_chk=pd.DataFrame(['buy_sell_hold_label']) #Block chain to store all the data from the classification labelling

    for ind,vals in df_series.iteritems():

        if vals>percent: #Classifying that signal denotes a buy past a certain threshold for labelling 
            temp='B'
            blk_chk=blk_chk.append({'buy_sell_hold_label':temp},ignore_index=True)
        elif vals <-1*percent: #Classifying that volatility denotes selling past a certain threshold for labelling 
            temp='S'
            blk_chk=blk_chk.append({'buy_sell_hold_label':temp},ignore_index=True)
        else:
            temp='H' #Determined that time series data is within threshold. and therefore is in a hold pattern. 
            blk_chk=blk_chk.append({'buy_sell_hold_label':temp},ignore_index=True) #Classifying that volatility denotes a hold within percent threshold specified by user  

    blk_chk=blk_chk.iloc[1:,1]

    blk_chk.index=df.index
    df=pd.concat([df, blk_chk],axis=1)#concatanation of the original dataframe to new classification labels

    return df #Returning original Dataframe with nre information 


def stat_significant(row,threshold=0.05): #Take input of financials statisitical result and returns a True/False string as to whether or not it is a stastical result is true. 
    if row<threshold:
        ret_val='True'
    else:
        ret_val='False'
    return ret_val
   
```

## Time Series Exploratory Data analysis and Data Pre processing 

### Time Series New feature generation 

log means returns feature was generated to measure of index volatility:

-This is done in order to divide the log of the Closing value of the present $F_t$ from the closing value of the day before $F_{t-1}$ as per the equation:

$log \frac{F_t}{F_{t-1}} $

The results from this equation are stored in the feature variable Log_returns for each dataset. 

Log returns are used for a number of reasons:
1. Log returns are additive whereas scaled are not. This makes this variable as a result more computationaly useful. 
2. Log returns are also symmetric with respect to gains and losses. By comparison percent gains as per the equation:

$100*( \frac{F_t}{F_{t-1}}-1) $

are biased in favor of gains.

Once the Log returns values where generated they where discretized into Buy,Sell and hold signals using the control_chart object and stored in the column 'buy_sell_hold_label' for each index.

The volatility returns where discretised into three categories based on EDA reviewing the annual standard deviation being approx. 1% for each index:
- Buy categories for volatility >1%
- Sell category for volatility >-1%
- Hold for volatility <1% and <-1% 


```python
#Generation of new features for DJIA index 
DJIA_First_price=DJIA.Dow_Jones_Industrial_Average[0]
DJIA['DJIA_Shifted']=DJIA.Dow_Jones_Industrial_Average.shift() #Shifted value for log returns feature
DJIA['DJIA_cumalitive_percent_increase']= DJIA.Dow_Jones_Industrial_Average.div(DJIA_First_price).mul(100) #Cumalative returns calculated to determine the overall trend in performance of the index 
DJIA['pct_change'] = DJIA.Dow_Jones_Industrial_Average.pct_change()
DJIA['Log_return']=np.log(DJIA.Dow_Jones_Industrial_Average) - np.log(DJIA.DJIA_Shifted) #Log return of daily returns generated here. 
DJIA.Log_return=DJIA.Log_return.dropna()
DJIA=control_chart(DJIA,DJIA.Log_return,0.01)

#Generation of new features for ASX_200 index 
ASX_200_first_price=ASX_200.ASX_200_close_value[0]
ASX_200['ASX_200_Shifted']=ASX_200.ASX_200_close_value.shift()#Shifted value for log returns feature
ASX_200['ASX_200_cumalitive_percent_increase']= ASX_200.ASX_200_close_value.div(ASX_200_first_price).mul(100)
ASX_200['ASX_200_pct_change'] = ASX_200.ASX_200_close_value.pct_change()
ASX_200['ASX200_log_return']=np.log(ASX_200.ASX_200_close_value) - np.log(ASX_200.ASX_200_Shifted) #Log return of daily returns generated here.
ASX_200.ASX200_log_return=ASX_200.ASX200_log_return.dropna()
ASX_200=control_chart(ASX_200,ASX_200.ASX200_log_return,0.01)

#Generation of new features for ASX_200_financial sector 
ASX_200_financial_first_price=ASX_200_financials.ASX_Financials_close_value[0]
ASX_200_financials['ASX200_financial_Shifted']=ASX_200_financials.ASX_Financials_close_value.shift()#Shifted value for log returns feature
ASX_200_financials['ASX200_financial_cumalitive_percent_increase']= ASX_200_financials.ASX_Financials_close_value.div(DJIA_First_price).mul(100)
ASX_200_financials['ASX200_financial_pct_change'] = ASX_200_financials.ASX_Financials_close_value.pct_change()
ASX_200_financials['ASX200_financial_Log_return']=np.log(ASX_200_financials.ASX_Financials_close_value) - np.log(ASX_200_financials.ASX200_financial_Shifted) #Log return of daily returns generated here. 
ASX_200_financials.ASX200_financial_Log_return=ASX_200_financials.ASX200_financial_Log_return.dropna()
ASX_200_financials=control_chart(ASX_200_financials,ASX_200_financials.ASX200_financial_Log_return,0.01)
```

```python
from spacy.lang.en import English #importing english language varient of spacy for extraction of tokens from text data
parser = English()
def tokenize(text): #Tokenizing function which converts input 'text' string to spacy doc string list lda_tokens
    fin_tokens = [] #blank variable where final important token words from string text will be returned as the final string list
    tokens = parser(text) #converting text string to spacy doc doc 
    for token in tokens: #for loop iterating through words in string text input 
        if token.orth_.isspace(): #if space is found in text string ignore and continue for loop
            continue
        elif token.like_url: #function handle for recording url
            fin_tokens.append('URL')
        elif token.orth_.startswith('@'): #fundction handle for recording screen names of users
            fin_tokens.append('SCREEN_NAME')
        else:
            fin_tokens.append(token.lower_) #final else statement for recording all words in string into lda_tokens 
    return fin_tokens
```

```python
#Functions to be utilised in NMP analysis 


#Please note as well for these scripts to be utilised file directories and empty text files need to be generated for storing the following files:

#- Text_file
#- Corpus_file
#- Vocab_file
#- Results files W.txt,Wc,txt and H.txt which are used to store the final topic model results and the topic model itself. 

#A short description of what each object does can be found in the readme file

#Please note also the following:

#The maximum text file size that can be processed with 8GB of RAM  is <1800kB and that the average processing time for this file size with a modelling iterations setpoint of 100 is in excess of 1hour. 


def read_docs(file_name):
    print('read documents')
    print('-'*50)
    docs = []
    fp = open(file_name, 'r')
    for line in fp:
        arr = re.split('\s', line[:-1])
        arr = filter(None, arr)
        arr = [int(idx) for idx in arr]
        docs.append(arr)
    fp.close()
    
    return docs

def read_vocab(file_name):
    print('read vocabulary')
    print('-'*50)
    vocab = []
    fp = open(file_name, 'r')
    for line in fp:
        arr = re.split('\s', line[:-1])
        vocab.append(arr[0])
    fp.close()

    return vocab

def calculate_PMI(AA, topKeywordsIndex):
    '''
    Reference:
    Short and Sparse Text Topic Modeling via Self-Aggregation
    '''
    D1 = np.sum(AA)
    n_tp = len(topKeywordsIndex)
    PMI = []
    for index1 in topKeywordsIndex:
        for index2 in topKeywordsIndex:
            if index2 < index1:
                if AA[index1, index2] == 0:
                    PMI.append(0.0)
                else:
                    C1 = np.sum(AA[index1])
                    C2 = np.sum(AA[index2])
                    PMI.append(np.log(AA[index1,index2]*D1/C1/C2))
    avg_PMI = 2.0*np.sum(PMI)/float(n_tp)/(float(n_tp)-1.0)

    return avg_PMI

```

```python
'''
Short Text Topic Modeling via SeaNMF
'''
class SeaNMFL1(object):
    def __init__(
        self,
        A, S, 
        IW1=[], IW2=[], IH=[],
        alpha=1.0, beta=0.1, n_topic=10, max_iter=100, max_err=1e-3, 
        rand_init=True, fix_seed=False):
        '''
        0.5*||A-WH^T||_F^2+0.5*alpha*||S-WW_c^T||_F^2+0.5*beta*||W||_1^2
        W = W1
        Wc = W2
        '''
        if fix_seed: 
            np.random.seed(0)

        self.A = A
        self.S = S

        self.n_row = A.shape[0]
        self.n_col = A.shape[1]

        self.n_topic = n_topic
        self.max_iter = max_iter
        self.alpha = alpha
        self.beta = beta
        self.B = np.ones([self.n_topic,1])
        self.max_err = max_err

        if rand_init:
            self.nmf_init_rand()
        else:
            self.nmf_init(IW1, IW2, IH)
        self.nmf_iter()

    def nmf_init_rand(self):
        self.W1 = np.random.random((self.n_row, self.n_topic))
        self.W2 = np.random.random((self.n_row, self.n_topic))
        self.H = np.random.random((self.n_col, self.n_topic))

        for k in range(self.n_topic):
            self.W1[:, k] /= norm(self.W1[:, k])
            self.W2[:, k] /= norm(self.W2[:, k])

    def nmf_init(self, IW1, IW2, IH):
        self.W1 = IW1
        self.W2 = IW2
        self.H = IH

        for k in range(self.n_topic):
            self.W1[:, k] /= norm(self.W1[:, k])
            self.W2[:, k] /= norm(self.W2[:, k])

    def nmf_iter(self):
        loss_old = 1e20
        print('loop begin')
        start_time = time.time()
        for i in range(self.max_iter):
            self.nmf_solver()
            loss = self.nmf_loss()
            if loss_old-loss < self.max_err:
                break
            loss_old = loss
            end_time = time.time()
            print('Step={}, Loss={}, Time={}s'.format(i, loss, end_time-start_time))

    def nmf_solver(self):
        '''
        using BCD framework
        '''
        epss = 1e-20
        # Update W1
        AH = np.dot(self.A, self.H)
        SW2 = np.dot(self.S, self.W2)
        HtH = np.dot(self.H.T, self.H)
        W2tW2 = np.dot(self.W2.T, self.W2)
        W11 = self.W1.dot(self.B)

        for k in range(self.n_topic):
            num0 = HtH[k,k]*self.W1[:,k] + self.alpha*W2tW2[k,k]*self.W1[:,k]
            num1 = AH[:,k] + self.alpha*SW2[:,k]
            num2 = np.dot(self.W1, HtH[:,k]) + self.alpha*np.dot(self.W1, W2tW2[:,k]) + self.beta*W11[0]
            self.W1[:,k] = num0 + num1 - num2
            self.W1[:,k] = np.maximum(self.W1[:,k], epss) # project > 0
            self.W1[:,k] /= norm(self.W1[:,k]) + epss # normalize
        # Update W2
        W1tW1 = self.W1.T.dot(self.W1)
        StW1 = np.dot(self.S, self.W1)
        for k in range(self.n_topic):
            self.W2[:,k] = self.W2[:,k] + StW1[:,k] - np.dot(self.W2, W1tW1[:,k])
            self.W2[:,k] = np.maximum(self.W2[:,k], epss)
        #Update H
        AtW1 = np.dot(self.A.T, self.W1)
        for k in range(self.n_topic):
            self.H[:,k] = self.H[:,k] + AtW1[:,k] - np.dot(self.H, W1tW1[:,k])
            self.H[:,k] = np.maximum(self.H[:,k], epss) 

    def nmf_loss(self):
        '''
        Calculate loss
        '''
        loss = norm(self.A - np.dot(self.W1, np.transpose(self.H)), 'fro')**2/2.0
        if self.alpha > 0:
            loss += self.alpha*norm(np.dot(self.W1, np.transpose(self.W2))-self.S, 'fro')**2/2.0
        if self.beta > 0:
            loss += self.beta*norm(self.W1, 1)**2/2.0
        
        return loss
    
    def get_lowrank_matrix(self):
        return self.W1, self.W2, self.H
    
    def save_format(self, W1file='W.txt', W2file='Wc.txt', Hfile='H.txt'):
        np.savetxt(W1file, self.W1)
        np.savetxt(W2file, self.W2)
        np.savetxt(Hfile, self.H)
        
'''
Topic Modeling via NMF
'''
class NMF(object):
    def __init__(
        self, 
        A, IW=[], IH=[],
        n_topic=10, max_iter=100, max_err=1e-3,
        rand_init=True):
        '''
        A = WH^T
        '''
        self.A = A
        self.n_row = A.shape[0]
        self.n_col = A.shape[1]
        
        self.n_topic = n_topic
        self.max_iter = max_iter
        self.max_err = max_err

        self.obj = []
        if rand_init:
            self.nmf_init_rand()
        else:
            self.nmf_init(IW, IH)
        self.nmf_iter()
    
    def nmf_init_rand(self):
        self.W = np.random.random((self.n_row, self.n_topic))
        self.H = np.random.random((self.n_col, self.n_topic))

        for k in range(self.n_topic):
            self.W[:,k] /= norm(self.W[:,k])
                
    def nmf_init(self, IW, IH):
        self.W = IW
        self.H = IH

        for k in range(self.n_topic):
            self.W[:,k] /= norm(self.W[:,k])

    def nmf_iter(self):
        loss_old = 1e20
        print('loop begin')
        start_time = time.time()
        for i in range(self.max_iter):
            self.nmf_solver()
            loss = self.nmf_loss()
            self.obj.append(loss)

            if loss_old-loss < self.max_err:
                break
            loss_old = loss
            end_time = time.time()
            print('Step={}, Loss={}, Time={}s'.format(i, loss, end_time-start_time))
        print('loop end')
                
    def nmf_solver(self):
        '''
        regular NMF without constraint.
        Block Coordinate Decent
        '''
        epss = 1e-20
        
        HtH = self.H.T.dot(self.H)
        AH = self.A.dot(self.H)
        for k in range(self.n_topic):
            tmpW = self.W[:,k]*HtH[k,k] + AH[:,k] - np.dot(self.W, HtH[:,k])
            self.W[:,k] = np.maximum(tmpW, epss)
            self.W[:,k] /= norm(self.W[:,k]) + epss
        
        WtW = self.W.T.dot(self.W)
        AtW = self.A.T.dot(self.W)
        for k in range(self.n_topic):
            self.H[:,k] = self.H[:,k]*WtW[k,k] + AtW[:,k] - np.dot(self.H, WtW[:,k])
            self.H[:,k] = np.maximum(self.H[:,k], epss)

    def nmf_loss(self):
        loss = norm(self.A - np.dot(self.W, np.transpose(self.H)), 'fro')**2/2.0
        return loss
    
    def get_loss(self):
        return np.array(self.obj)
    
    def get_lowrank_matrix(self):
        return self.W, self.H
    
    def save_format(self, Wfile='W.txt', Hfile='H.txt'):
        np.savetxt(Wfile, self.W)
        np.savetxt(Hfile, self.H)
```

```python
def NMF_training_model(corpus_file_loc,vocab_file_loc,max_iter,n_topics):
    
    '''
    SeaNMF Training
    '''


    #parser = argparse.ArgumentParser()
    #parser.add_argument('--corpus_file', default='data/doc_term_mat.txt', help='term document matrix file')
    #parser.add_argument('--vocab_file', default='data/vocab.txt', help='vocab file')
    #parser.add_argument('--model', default='seanmf', help='nmf | seanmf')
    #parser.add_argument('--max_iter', type=int, default=500, help='max number of iterations')
    #parser.add_argument('--n_topics', type=int, default=100, help='number of topics')
    #parser.add_argument('--alpha', type=float, default=0.1, help='alpha')
    #parser.add_argument('--beta', type=float, default=0.0, help='beta')
    #parser.add_argument('--max_err', type=float, default=0.1, help='stop criterion')
    #parser.add_argument('--fix_seed', type=bool, default=True, help='set random seed 0')
    #args = parser.parse_args()

    args=easydict.EasyDict({'corpus_file':corpus_file_loc, #'data/doc_term_mat_abcnews.txt', #Loading model attributes into a single file 
                            'vocab_file':vocab_file_loc,#'data/vocab.txt',
                            'model':'seanmf',
                            'max_iter':max_iter,#100,
                            'n_topics':n_topics,#300,
                            'alpha':0.1,
                            'beta':0.0,
                            'max_err':0.1,
                            'fix_seed':True
                            })

    docs = read_docs(args.corpus_file) #Reading in text corpus file 
    vocab = read_vocab(args.vocab_file) #Reading in vocabulary file
    n_docs = len(docs) #Determining the total number of documents 
    n_terms = len(vocab) #Determining the total number of terms
    print('n_docs={}, n_terms={}'.format(n_docs, n_terms))

    tmp_folder = 'seanmf_results'
    if not os.access(tmp_folder, os.F_OK): #accesining folder seanmf_results to store final results in text format in
        os.mkdir(tmp_folder)

    if args.model.lower() == 'nmf': 
        print('read term doc matrix')
        dt_mat = np.zeros([n_terms, n_docs]) #initialising empty term document matrix to be filled during processing
        for k in range(n_docs):
            for j in docs[k]: # filling counts of each word in term document matrix 
                dt_mat[j, k] += 1.0
        print('term doc matrix done')
        print('-'*50)

        model = NMF( #generating NMF model with parameters term document matrix, max error, max number of toics and max number of iterations
            dt_mat, 
            n_topic=args.n_topics, 
            max_iter=args.max_iter, 
            max_err=args.max_err)

        model.save_format( #Save model in temporary folder
            Wfile=tmp_folder+'/W.txt',
            Hfile=tmp_folder+'/H.txt')

    if args.model.lower() == 'seanmf': #Generate co-occurance matrix 
        print('calculate co-occurance matrix')
        dt_mat = np.zeros([n_terms, n_terms]) #generate empty co-occurence to store data into. 
        for itm in docs:
            for kk in itm:
                for jj in itm:
                    dt_mat[int(kk),int(jj)] += 1.0 #Adding value each time cooccurence is denoted between words 
        print('co-occur done')
        print('-'*50)
        print('calculate PPMI')
        D1 = np.sum(dt_mat) #PMI value calculated between words in topic to determine the give an indicator of cooccurence of these values 
        SS = D1*dt_mat
        for k in range(n_terms):
            SS[k] /= np.sum(dt_mat[k])
        for k in range(n_terms):
            SS[:,k] /= np.sum(dt_mat[:,k])
        dt_mat = [] # release memory
        SS[SS==0] = 1.0
        SS = np.log(SS)
        SS[SS<0.0] = 0.0
        print('PPMI done')
        print('-'*50)

        print('read term doc matrix') #Size of term document matrix primary cause for memory error please ensure text corpus<1800kB
        dt_mat = np.zeros([n_terms, n_docs])
        for k in range(n_docs):
            for j in docs[k]: #Fill in term document matrix into a numpy array 
                dt_mat[j, k] += 1.0
        print('term doc matrix done')
        print('-'*50)

        model = SeaNMFL1(
            dt_mat, SS,  
            alpha=args.alpha, 
            beta=args.beta, 
            n_topic=args.n_topics, 
            max_iter=args.max_iter, 
            max_err=args.max_err,
            fix_seed=args.fix_seed)

        model.save_format(
            W1file=tmp_folder+'/W.txt',
            W2file=tmp_folder+'/Wc.txt',
            Hfile=tmp_folder+'/H.txt')

```

## LDA topic modelling 

LDA Topic modelling of this nature was applied over the following time intervals:

-  2008 to 2011 

The initial range of topics where set between 25 and 50 for the time period from 2008 to 2011. The topic range was based on historical topic modelling of news articles [1]. All of the models topics generated where assessed on semantic coherence the following criteria:

1.	Manual Human User assessment for semantic coherence of the document topics
2.	Semantic coherence testing using CV semantic topic coherence which has been found to be correlate best in the literature with qualitative human interpretation of semantic topic coherence[2] 

This was done in order to determine if the Cv semantic topic coherence score could be utilised for assessing topic coherence for short text corpus.Performance ws found to be poor initially (coherence <30%) and was denoted to more than likely due to the term document matrix being very sparse. The hyperparameters of the model where tuned further to see if performance could be improved. These tuning approaches where:

1.	Increasing the number of topics up to a maximum of 400. This was done so as a stop check just to confirm the incoherent vague topics was not as a result of not enough topics being specified which is another typical problem that can be encountered with LDA topic modelling[3].
2.	The headline document text corpus was aggregated by date to increase the text corpus size (which is suggested in the literature as way to overcome this issue [4]). 
3.	Changing the dictionary from a Bag of Words dictionary to TF-IDF dictionary to remove infrequent or very frequent keywords for the topic modelling. This was done to reduce down the overall text sparsity in the term document matrix by reducing the overall size of the text corpus. 

Topic semantic coherence improved by approximately 20% with the TF-IDF model albeit and data aggregation no real improvement in topic coherence was found with the increasing the topic number approaches or they became worse(See figure below this cells). Hence another novel topic model approach specifically designed to handle short text documents and this sparsity problem proposed by Yan [5] was utilised instead. A further description of this model can be found in the NMF modelling section. 

-  [1]W. Jacobi, Quantitative analysis of large amounts of journalistic texts using topic modelling, Digital Journalism, vol. 4, no. 1, pp. 89-106, 2016. 
-  [2]Michael Rder, Exploring the Space of Topic Coherence Measures, in Proceedings of the eighth ACM international conference on Web search and data mining. , New York, 2015. 
-  [3]Chae, J., Thom, D., Bosch, H., Jang, Y., Maciejewski, R., Ebert, D.S. and Ertl, T., 2012, October. In Visual Analytics Science and Technology (VAST), 2012 IEEE Conference on (pp. 143-152). IEEE.

# NMF model generation 

The NMF (Non-negative Matrix Factorisation) model selected to resolve the term -document matrix sparsity issue utilises a term correlation matrix instead of a term document matrix that is utilised with the LDA model. This term correlation matrix is a dense matrix and hence overcomes the sparsity issue as a result. This NMF models in essence identifies topics based on the idea that most topics are uncovered based on correlations between terms. An unusual phenomenon that is observed with short text documents is that despite how large the text corpus grows in size the number of distinct terms remains relatively static reaching a maximum ratio of unique terms to overall corpus size of 1:3 once the number of text documents in >10000(See Figure in cell below). A similar behaviour was observed with the text corpus used in this module of 184:1 with a corpus size of 6.6x106 words(Reference EDA section of this notebook). For further information as to how this model calculated the term correlation matrix please refer to [1].
 
Due to space complexity topic modelling was applied to sub sampled headline text corpus from 01-Jan to 30-Jun of the years to 2009, 2012 and 2014. The parameterisation grid search of the topic model was done with the 2009 data set. As the topic modelling was being done with respect to a term correlation matrix Positive Point Mutual Information (PPMI) was utilised instead as is suggested by Yan. 

A human qualitative assessment of the 20 topics with the lowest PMI topic was assessed for topic coherence also to confirm the effectiveness of this approach and to determine parameterisation of what the optimum number of topics to be selected for the model should be. From this analysis the optimum number of topics for each text corpus was found to be 200 as it yielded the highest average PMI score across all the topics generated along with the bottom 20 topics retaining an average PMI score of 0.89(See table below) with 95% of the topics verified through human classification as coherent also. 

As a result topic modelling was performed across all 3 text time points in 2009,2012 and 2014. All topics generated where reviewed manually for coherence and to see if further aggregation of the different topics was feasible. From this review it was found that it could be specifically topics relating to cricket, soccer etc could be related aggregated into a sport topic etc. This was done for each topic. As a result of the total number of unique topic labels for each data set was reduced by approx 40%. The results for this cannot be shown here dues to size constraints but can be found in the appendices results file attached to this notebook.     

-  [1]W. Yan Xiaohui, Learning Topics in Short Texts by Non-negative Matrix Factorization on term correlation matrix, in Proceedings of the 2013 SIAM International Conference on Data Mining, Austin Texas, 2013.

## Topic assignment of headlines documents for analysis 

Once the sampling of the topic modelling of the three time points was complete topic assignment to each document was performed. The major assumption made for this approach was that only one topic was assigned to each document. This assumption was justified being that the average size of the document being <80 characters the odds of more than one major topic being present was low. The methodology for assigning a topic to each document was the following:

1.	All topics keywords are cross checked against the words present in the document string. 
2.	If a topic keyword is present in the topic string its associated PMI score for that topic is cumulatively assigned to that topic for that document for each keyword present. 
3.	The topic assigned to that document is then the determined by the highest cumalitive PMI score. 

The justification for this approach is that, with only one topic being assigned to each document the topic which has the highest number of keywords with the greatest likelihood of having other keywords within that topic being present(which is what PMI estimates) in the document is the most likely topic of that document. 
Once the initial topic classification was complete these topics where compared against the aggregate topic labels to further simplify and reduce down the number of topics present in the document classification dataset. Any document that could not be classified by this approach was discarded from further analysis.

### Object oriented programs used in topic classification

```python
#Assigning different topics to different news headlines in 2009 jan to jun dataframe based on maximum liklikhood principle 

from collections import defaultdict

#Code to assign different topics to each string in the dataset based on keywords present in the string. 

def contains_word(s, w): #Combining split strings 
    return (' ' + w + ' ') in (' ' + s + ' ')

def document_topic_classifier(df_headlines,df_topic_list,Topic_agg=0):
    
    topic_blk=[] #Initialising blank list where topic will be stored. 
    temp_str='No topic assignment'
    temp_str2='Short_topic_aggregated'
    
    
    topics_list=[] #Splitting dataframe string in to words to iterate through 

    for idx, vals in df_topic_list.Topic_string.iteritems():
        topics_list.append(vals)

    topic_year_split_list=[]

    for vals in topics_list: #iterating through each topic spliting string to iterate through against headline
        ar_temp=0
        ar_temp=vals.split()
        topic_year_split_list.append(ar_temp)
    
    for idx,vals in df_headlines.headline_text.iteritems(): #iterating through each index headlines

        #print(f'headlines is: {vals}')
        blk_dict=defaultdict(int) #Initialing default dictionary to store data in
        ar_temp=0
        for topics in topic_year_split_list: #For loop iterating through each topic 
            #iterating through each topic 
            key=0
            key=' '.join(topics)
            
            for words in topics: #iterating through each word in topic 

                if contains_word(vals,words)==True: #comparing words in topic to words in headline if present 
                    ar_temp2=0
                    ar_temp2=df_topic_list.loc[df_topic_list['Topic_string'] == key, 'PMI_value'] #Acquiring PMI value from topic dataframe
                    #print(f'key is:{key}')
                    blk_dict[key]+=float(ar_temp2.item()) #Assigning PMI value for word to dataframe
                    
                    #print(f'dictionary is {blk_dict}')
                    #If word is present topic value increases by one. 
        

        if len(blk_dict)>0:
    
            ar_temp=max(blk_dict, key=blk_dict.get)   #Assign topic to maximum value pair 
            
            #print(f'max value pair is {ar_temp}')
            topic_blk.append(ar_temp)
        else:
            topic_blk.append(temp_str)
        
    if Topic_agg==True:    #confirm topic aggregation

        final_blk=[]#Reassign topics of list to topic aggregation labels if true 

        Agg_dict=pd.Series(df_topic_list.Short_topic_aggregated.values,index=df_topic_list.Topic_string).to_dict()#Creating dictoinary of topics and agg topic to iterate through

        for long_topics in topic_blk: #Iterating through block of long topics and comparing to unique key of values. 
            
            if long_topics==temp_str: #No aggregation topic present still assign 'No topic assignment' for filtering of unclassified values at a later point. 
                final_blk.append(temp_str)
                
            else: #If topic is classified proceed to iterate and compare long topic to shorten down aggregated topic. 
                
                for key,value in Agg_dict.items():
                    
                    if key == long_topics:
                                                #if key present in topic _blk append said value of key to final_blk 
                        final_blk.append(value)
                               
    else: #If topic aggregation not required move to return final value as final_blk

        final_blk=topic_blk
               
    return final_blk
        
```

```python
from scipy.stats import chi2_contingency

def chisq_of_df_cols(df, c1, c2):
    groupsizes = df.groupby([c1, c2]).size()
    ctsum = groupsizes.unstack(c1)
    # fillna(0) is necessary to remove any NAs which will cause exceptions
    return(chi2_contingency(ctsum.fillna(0)))

def general_BSH_ratios(df):
    
    cr1=df.DJIA_BuySellHoldLabels.value_counts()
    
    cr1_BHR=cr1.iloc[2]/cr1.iloc[0]
    cr1_SHR=cr1.iloc[1]/cr1.iloc[0]

    cr2=df.ASX_200_BuySellHoldLabels.value_counts()
 
    cr2_BHR=cr2.iloc[2]/cr2.iloc[0]
    cr2_SHR=cr2.iloc[1]/cr2.iloc[0]
    
    cr3=df.ASX_200_Fin_BSHLabels.value_counts()
    
    cr3_BHR=cr3.iloc[2]/cr3.iloc[0]
    cr3_SHR=cr3.iloc[1]/cr3.iloc[0]
    
    l_BHR=[cr1_BHR, cr2_BHR,cr3_BHR]

    l_SHR=[cr1_SHR, cr2_SHR,cr3_SHR]
    
    df_BSH_summary=pd.concat([cr1, cr2,cr3], axis=1)
    df_BSH_summary_list=df_BSH_summary.values.tolist()
    df_BSH_summary_list.append(l_BHR)
    df_BSH_summary_list.append(l_SHR)
    
    
    index_list=['B','H','S','BHR','SHR']
    
    col_names =['DJIA_BuySellHoldLabels','ASX_200_BuySellHoldLabels','ASX_200_Fin_BSHLabels']
    
    df_final_sum=pd.DataFrame(df_BSH_summary_list,
                          index=index_list,
                          columns=['DJIA_BuySellHoldLabels','ASX_200_BuySellHoldLabels','ASX_200_Fin_BSHLabels'])
    
    return df_final_sum

 
def generate_contigencyplot(drop_col_list,Gen_df_pre_drop,groupby_string,label_tr):
#Generate new pivot contigency table

    DJIA_2009_BSH=Gen_df_pre_drop.drop(drop_col_list,axis=1) #drop indexes from columns that are not being assessed 
    count_temp=DJIA_2009_BSH.groupby(['topic_assigned',groupby_string]).count() #Count parameters from 

    New_set=count_temp.reset_index().pivot('topic_assigned',groupby_string,'headline_text')
    New_set['Buy:Hold']=New_set['B']/New_set['H']
    New_set['Sell:Hold']=New_set['S']/New_set['H']

    #Generation of threshold markers to delineate categorical variables that are to be used as features
    New_set_mean_buy=New_set['B'].mean()
    New_set_mean_sell=New_set['S'].mean()
    New_set_mean_hold=New_set['H'].mean()

    New_set_mean_buy_hold_ratio=New_set_mean_buy/New_set_mean_hold
    New_set_mean_sell_hold_ratio=New_set_mean_sell/New_set_mean_hold

    #Generating general plot to observe all buy sell and hold signals for time period across all buy sell and hold quadrants 

    ax=New_set.plot(kind='scatter',y='Buy:Hold',x='Sell:Hold',figsize=(20,10)) #threshold set at 0.5 
    ax.axhline(New_set_mean_buy_hold_ratio,color='g',label='buy threshold')
    ax.axvline(New_set_mean_sell_hold_ratio,color='r',label='sell threshold')

    plt.text(New_set_mean_sell_hold_ratio+label_tr, #Label threshold inputted to ensure they can be realigned for different scalings of graph
             New_set_mean_buy_hold_ratio+label_tr,
             'Highly volatile',fontsize=12,
             bbox=dict(facecolor='orange', alpha=0.5))
    
    plt.text(New_set_mean_sell_hold_ratio-label_tr, 
             New_set_mean_buy_hold_ratio+label_tr,
             'Buy signals',
             fontsize=12,
             bbox=dict(facecolor='green', alpha=0.5))
    
    plt.text(New_set_mean_sell_hold_ratio+label_tr,
             New_set_mean_buy_hold_ratio-label_tr,
             'Sell Signals',
             fontsize=12,
             bbox=dict(facecolor='red', alpha=0.5))
    
    plt.text(New_set_mean_sell_hold_ratio-label_tr,
             New_set_mean_buy_hold_ratio-label_tr,
             'Lower Volatility',
             fontsize=12,
             bbox=dict(facecolor='grey', alpha=0.5))
    plt.title('General buy,sell and hold plot',fontsize=12)
    plt.legend() 
    fig1 = plt.gcf()
    #fig1.set_size_inches(11,8)
    fig1.savefig('temp_mainchart.png') #Saving plot to file 
    
    plt.show()
    

   
    
    #Creating new arrays for dataset for visualising threshold data quadrants 
    #High volatility labels  Chart generation
    BSH_HV_BVlabel=New_set.loc[(New_set['Buy:Hold']>=New_set_mean_buy_hold_ratio)&
                               (New_set['Sell:Hold']>=New_set_mean_sell_hold_ratio),
                               'Buy:Hold']
    BSH_HV_SVlabel=New_set.loc[(New_set['Buy:Hold']>=New_set_mean_buy_hold_ratio)&
                               (New_set['Sell:Hold']>=New_set_mean_sell_hold_ratio),
                               'Sell:Hold']

    fig, ax = plt.subplots(figsize=(20,10))
    ax.scatter(BSH_HV_SVlabel, BSH_HV_BVlabel,color='orange')

    for i,vals in enumerate(BSH_HV_BVlabel.index): #Annotating and labelling all points in the chart with news topics
        ax.text(BSH_HV_SVlabel.iloc[i],
                BSH_HV_BVlabel.iloc[i],
                vals)

    fig2 = plt.gcf() #Saving plot to file
    #fig1.set_size_inches(11,8)
    fig2.savefig('high_volatility_chart.png')    
        
    plt.show()


    #Low volatility labels  Chart generation
    BSH_LV_BVlabel=New_set.loc[(New_set['Buy:Hold']<=New_set_mean_buy_hold_ratio)&
                               (New_set['Sell:Hold']<=New_set_mean_sell_hold_ratio),
                               'Buy:Hold']

    BSH_LV_SVlabel=New_set.loc[(New_set['Buy:Hold']<=New_set_mean_buy_hold_ratio)&
                               (New_set['Sell:Hold']<=New_set_mean_sell_hold_ratio),
                               'Sell:Hold']

    fig, ax = plt.subplots(figsize=(20,10))
    ax.scatter(BSH_LV_SVlabel, BSH_LV_BVlabel,color='grey')

    for i,vals in enumerate(BSH_LV_BVlabel.index):
        ax.text(BSH_LV_SVlabel.iloc[i],
                BSH_LV_BVlabel.iloc[i],
                vals)
        
        
    fig3 = plt.gcf() #Saving plot to file
    #fig1.set_size_inches(11,8)
    fig3.savefig('low_volatility_chart.png')
    plt.show()
    

    #Buy Signals Chart generation
    BSH_BS_BVlabel=New_set.loc[(New_set['Buy:Hold']>New_set_mean_buy_hold_ratio)&
                               (New_set['Sell:Hold']<New_set_mean_sell_hold_ratio),
                               'Buy:Hold']
    BSH_BS_SVlabel=New_set.loc[(New_set['Buy:Hold']>New_set_mean_buy_hold_ratio)&
                               (New_set['Sell:Hold']<New_set_mean_sell_hold_ratio),
                               'Sell:Hold']

    fig, ax = plt.subplots(figsize=(20,10))
    ax.scatter(BSH_BS_SVlabel, BSH_BS_BVlabel,color='green')

    for i,vals in enumerate(BSH_BS_BVlabel.index):
        ax.text(BSH_BS_SVlabel.iloc[i],
                BSH_BS_BVlabel.iloc[i],
                vals)

    fig4 = plt.gcf()
    #fig1.set_size_inches(11,8)
    fig4.savefig('Buys_Signal_chart.png')
    plt.show()


    #Sell Signals Chart generation 
    BSH_SS_BVlabel=New_set.loc[(New_set['Buy:Hold']<New_set_mean_buy_hold_ratio)&
                               (New_set['Sell:Hold']>New_set_mean_sell_hold_ratio),
                               'Buy:Hold']
    BSH_SS_SVlabel=New_set.loc[(New_set['Buy:Hold']<New_set_mean_buy_hold_ratio)&
                               (New_set['Sell:Hold']>New_set_mean_sell_hold_ratio),
                               'Sell:Hold']

    fig, ax = plt.subplots(figsize=(20,10))
    ax.scatter(BSH_SS_SVlabel, BSH_SS_BVlabel,color='red')

    for i,vals in enumerate(BSH_SS_BVlabel.index):
        ax.text(BSH_SS_SVlabel.iloc[i],
                BSH_SS_BVlabel.iloc[i],
                vals)
        
        
    fig5 = plt.gcf()
    #fig1.set_size_inches(11,8)
    fig5.savefig('Sell_Signal_chart.png')

    plt.show()


```

```python
#Group slice and chop datase into an effective grouping to be utilised in object for SVM analysis

drop_col_list=['ASX_200_BuySellHoldLabels','ASX_200_Fin_BSHLabels','headline_text']
col_analysed='DJIA_BuySellHoldLabels'
df_BSH=topic_list_2009_BSH

#Purpose of this function is to generate a dataframe that can be used in the SVM machine learning algorithm 
#it reshapes the previous dataframes used in to a set of topic features counts and the volatility score is used as a label.  that are used to 

def SVM_create_count_table(drop_col_list,col_analysed,df_BSH):

    drop_df=df_BSH.drop(drop_col_list,axis=1) #drop indexes from columns that are not being assessed 

    count_temp=drop_df.groupby(['Date','topic_assigned']).count() #Count parameters from group by buy sell and hold values to create contigency table 

    New_set=count_temp.reset_index().pivot(index='Date', columns='topic_assigned',values=col_analysed) 

    #Filling in blank variables with 0 for analysis so SVM does not through error on NAN value
    New_set.fillna(value=0,inplace=True)

    clt=df_BSH.loc[:,['Date',col_analysed]] #Selecting panda series for concatanating to reindexed table
    clt.set_index('Date',inplace=True)

    clt = clt[~clt.index.duplicated(keep='first')] #Remove duplicate values for BSH prior to adding to dataset

    final_df=pd.concat([New_set,clt],axis=1) #Combine data set into final dataframe for processing in file
    
    return final_df

RSVM=SVM_create_count_table(drop_col_list,col_analysed,df_BSH)

RSVM.head()
```
